Import of library for data cleaning

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

Read csv file and added three column (year,month and day) for further sorting of data

```python
air_df = pd.read_csv('C:\\Users\\User\\Documents\\APIMS-final.csv')
air_df['Time'] = pd.to_datetime(air_df.Time)
air_df['year'] = pd.DatetimeIndex(air_df.Time).year
air_df['month'] = pd.DatetimeIndex(air_df.Time).month
air_df['day'] = pd.DatetimeIndex(air_df.Time).day
air_df
```

Extraction of data from 2018 to 2022

```python
clean_ap=pd.DataFrame()
for x in range(2018,2023,+1):
    air_want_df=air_df.loc[air_df.year==x]
    clean_ap=pd.concat([clean_ap, air_want_df])
clean_ap.reset_index(drop=True, inplace=True)
clean_ap
```

Identify places with high percentage of missing value which cant be filled by any method

```python
missing_value_df = clean_ap.isnull().sum()
raw_percent_missing_value = 100 * (missing_value_df / len(clean_ap.index))
percent_missing_value = round(raw_percent_missing_value, 2)
percent_missing_value.sort_values(ascending=False)
```

Muar with 95.96% of missing values is removed

```python
clean_ap.drop('Muar', axis=1, inplace=True)
```

Imputation of null values in a data frame involves filling in missing values in the data set. In this particular case, the Air Pollution Index (API) was taken at one-hour intervals, so the API values for adjacent time periods are expected to be similar. Therefore, forward-fill and back-fill methods can be used to fill in missing values for short periods of time. For long periods of missing data, fill in the missing values with the mean of the surrounding values.
1. Backfill
2. Forward fill
3. Fill with mean

```python
for col in clean_ap.columns:
    clean_ap[col].fillna(method='bfill', limit=90, inplace=True)

for col in clean_ap.columns:
    clean_ap[col].fillna(method='ffill', limit=90, inplace=True)

for col in clean_ap.columns:
    if col!='Time' and col != 'year' and col != 'day' and col != 'month':  # Skip the 'Year', 'Day', and 'Month' columns
        for year in clean_ap['year'].unique():
            mean_api = clean_ap.loc[(clean_ap['year'] == year), col].mean()
            # Replace missing values with the mean API
            clean_ap[col].fillna(mean_api, inplace=True)
clean_ap
```

Verify that the imputation has been successful by checking for any remaining missing values

```python
missing_value_df = clean_ap.isnull().sum()
raw_percent_missing_value = 100 * (missing_value_df / len(clean_ap.index))
percent_missing_value = round(raw_percent_missing_value, 2)
percent_missing_value.sort_values(ascending=False)
```

Convert the clean_ap data frame to csv file for further data analysis 

```python
clean_ap.to_csv('cleaned_air_data.csv', index=False)
```
